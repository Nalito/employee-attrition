```python
df = df[["Age", "Attrition", "Education", "EnvironmentSatisfaction", "Gender", "HourlyRate", "JobSatisfaction", "MaritalStatus", "MonthlyIncome", "OverTime", "WorkLifeBalance", "PerformanceRating", "YearsAtCompany"]]
```

```python
df["Attrition"] = np.where(df["Attrition"]=="Yes", 1, 0)
df["Gender"] = np.where(df["Gender"] == "Male", 1, 2)
df["MaritalStatus"] = np.where(df["MaritalStatus"]=="Single", 1, np.where(df["MaritalStatus"]=="Married", 2, 3))
df["OverTime"] = np.where(df["OverTime"] == "Yes", 1, 0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix, roc_auc_score, roc_curve
```

```python
X = df.drop(["Attrition"], axis=1)
y = df['Attrition']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=3, test_size=0.3)
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
log = LogisticRegression()
log.fit(X_train_scaled, y_train)
accuracy_score(y_test, log.predict(X_test_scaled))
```

```python
svc = SVC()
svc.fit(X_train_scaled, y_train)
accuracy_score(y_test, svc.predict(X_test_scaled))
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train_scaled, y_train)
accuracy_score(y_test, dtc.predict(X_test_scaled))
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train_scaled, y_train)
accuracy_score(y_test, rfc.predict(X_test_scaled))
```

```python
param_grid = {
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],
    'C': np.logspace(-4, 4, 10),
    'solver': ['liblinear', 'newton-cg', 'lbfgs', 'sag', 'saga'],
    'max_iter': [100, 1000, 5000],
    'class_weight': [None, 'balanced']
}

log_ = LogisticRegression()
grid_search = GridSearchCV(log_, param_grid, cv=5, verbose=1, n_jobs=-1)

grid_search.fit(X_train_scaled, y_train)

print("Best parameters:", grid_search.best_params_)
print("Best cross-validation score:", grid_search.best_score_)
```

```python
test_score = grid_search.score(X_test, y_test)
```
